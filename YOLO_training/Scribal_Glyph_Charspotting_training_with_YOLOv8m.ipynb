# Project task 2: Training the model

```python
from ultralytics import YOLO

dataset_name = "/content/drive/MyDrive/dataset/scribal-glyph-charspotting.yaml"

# Initialize YOLO Model
model = YOLO("yolov8m.pt")

wandb.init()

model.train(
    project="/content/drive/MyDrive/dataset/ultralytics_yolov8m_runs",
    data=dataset_name,
    epochs=200,
    imgsz=512,
    patience = 50,
    fliplr = 0.0,
    name = "exp_train_project_task_2",
    amp = True)

model.val(
    project="/content/drive/MyDrive/dataset/ultralytics_yolov8m_runs",
    data=dataset_name,
    imgsz=512,
    batch=16,
    name="exp_train_project_task_2/val",
    plots=True

)

# Finish the W&B run
wandb.finish()
```

# Evaluating the results of task 2 training on the test set

```python
#!pip install --upgrade ultralytics wandb
import os, torch
import numpy as np
import warnings
warnings.filterwarnings("ignore")
os.environ["YOLO_VERBOSE"] = "True"
from ultralytics import YOLO

best_model = YOLO("/content/drive/MyDrive/dataset/ultralytics_yolov8m_runs/exp_train_project_task_2/weights/best.pt")

dataset_name = "/content/drive/MyDrive/dataset/scribal-glyph-charspotting.yaml"

# ---------------- Best Model Evaluation ----------------
test_metrics = best_model.val(
    data=dataset_name,
    project="/content/drive/MyDrive/dataset/ultralytics_yolov8m_runs/exp_train_project_task_2",
    split="test",
    name="test",
    imgsz=512,
    batch=16,
    plots=True,
    save=True
)

print("Test mAP50-95:", test_metrics.box.map)
print("Test mAP50:", test_metrics.box.map50)
print("Test Precision:", np.mean(test_metrics.box.p))
print("Test Recall:", np.mean(test_metrics.box.r))
```
